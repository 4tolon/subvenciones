# Ánalisis Exploratorio del csv de convocatorias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_context('poster')
sns.set(rc={'figure.figsize': (16., 9.)})
sns.set_style('whitegrid')
```

```python
#Carga de datos
df_co = pd.read_csv('../files/convocatorias.csv.gz', compression='gzip', low_memory=False)
```

```python
#Dimensiones originales
df_co.shape
```

```python
# info y nulos
df_co.info()
```

```python
# Parte final de los datos
df_co.tail()
```

```python
# La columna 5 tiene algo de informacion:
df_co['Unnamed: 5'].unique()[0:5]
```

```python
# La columna 9 tiene algo de informacion:
df_co['Unnamed: 9'].unique()[0:3]
```

```python
# Relleno los nan con '-'
df_co['Unnamed: 9'].fillna('-', inplace=True)
df_co['Unnamed: 5'].fillna('-', inplace=True)
```

```python
# Parece que las 3 úlimas columnas no tiene informacion, la quitamos:
df_co = df_co.iloc[: , :-3]
df_co.sample(5)
```

```python
# Traspongo, reseteo indice y vuelvo a tradponer para conservar primera fila
df_co = df_co.T.reset_index().T.reset_index(drop=True)
```

```python
df_co.head(3)
```

```python
# Cambiamos el los registros de la primera fila que hemos arrastrado del procesado de los datos.
df_co[5][0] = '-'
df_co[9][0] = '-'
```

```python
df_co.sample(5)
```

```python
# Renombramos las columnas en  funcion de las columnas que aparecen en los resultados de búsquedas de la web de convocatorias. La primer y úlima las nombro segun lo que me parece
columnas = ['Id_original', 'Código_BDNS', 'MMR', 'Administración', 'Departamento', 'Órgano', 'Fecha_de_registro', 'Título_de_la_convocatoria', 'URL', 'Extra']
df_co.set_axis(columnas, axis='columns', inplace=True)
```

```python
# Conversion de la columna fecha a Datetime
df_co['Fecha_de_registro']=pd.to_datetime(df_co['Fecha_de_registro'], format="%d/%m/%Y")
```
